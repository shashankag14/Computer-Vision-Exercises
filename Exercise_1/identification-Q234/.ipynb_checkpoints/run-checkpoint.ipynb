```python
D = np.arange(9).reshape(3,3) 

recall = []
precision = []
total_imgs = D.shape[1]

num_images = D.shape[0]
#assert(D.shape[0] == D.shape[1])

# Creates an identity matrix of size = num_images
labels = np.diag([1]*num_images)

# Flattens the D and labels array
d = D.reshape(D.size)
l = labels.reshape(labels.size)

sortidx = d.argsort()
d = d[sortidx]
l = l[sortidx]

tp = 0
fn = 0 
fp = 0
threshold = list(range(0, 5, 1))

for threshold_idx in threshold:
    for idx in range(len(d)):
        if (d[idx] < threshold_idx):
            tp = tp + l[idx]
            fp = np.sum(l[0:idx+1] == 0)
        else :
            fn = fn + l[idx]
    print(tp,fp,fn)

    # compute precision and recall values and append them to "recall" and "precision" vectors
    # your code here
    if (tp + fp == 0):
        precision.append(1)
    else:
        precision.append(tp/(tp+fp))
    if (tp + fn == 0):  
        recall.append(1)
    else:
        print(tp+fn)
        recall.append(tp/(tp+fn))
        
#print(d, l)
print(precision, recall)
```
